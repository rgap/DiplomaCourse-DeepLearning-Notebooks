##Gradiente descendente
En el siguiente ejercicio codificar치s el descenso por gradiente para ajustar los pesos sin치pticos de un perceptr칩n.

```
import numpy as np
```

```
def sigmoide(x):
  sig = 1/(1 + np.exp(-x))
  return sig
```

## Par치metros de entrada

```
x = np.array([0.3, -0.5, 0.7, 1])
y = 0.3
w = np.random.randn(4)
indiceAprendizaje = 0.7
```

## Algoritmo de aprendizaje

```
for epocas in range(150):
  h = np.dot(w, x)
  out = sigmoide(h)
  error = (y - out)
  error_term = error * sigmoide(h) * (1 - sigmoide(h))
  del_w = indiceAprendizaje * error_term * x
  w = w + del_w
  print('salida:', out)
  print('error:', error)
```
